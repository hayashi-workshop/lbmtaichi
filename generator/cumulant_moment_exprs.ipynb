# Central and raw moments

Raw moments in velocity distributions

Instantiate MomentGenerator for D3Q27

```python
moment_d3q27 = MomentGenerator(vectors=create_vectors(dim=3), is_central_moment=False)
```

Definition of raw moment

```python
moment_d3q27.derive_symbolic_moment_sum()
```

Moment as function of velocity distribution

```python
moment_d3q27((0,2,0))
```

Switch to Central Moment (kappa) mode

```python
moment_d3q27.is_central_moment = True
moment_d3q27.derive_symbolic_moment_sum()
```

[!NOTE] the first order central moments, (1,0,0) (0,1,0) (0,0,1), are identically zero, but they will show multiple terms here. They become zero if the velocity components are expressed with f. The code avoid this substitution since the computation can be very slow for high order moments. 

```python
moment_d3q27((1,1,0))
```

The central moment derived from raw moments. 

```python
moment_d3q27.derive_central_moment_from_moment((0,2,0))
```

Backward transformation from raw moment to distribution

```python
f_exprs, f_list = moment_d3q27.express_f_in_terms_of_moments()
```

```python
q = 1
lhs = sp.Symbol(f'f_{q}')
display(Eq(lhs, f_exprs[0]))
```
